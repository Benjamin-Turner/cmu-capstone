```python
# 1. Remove rows
start_rows = len(cleaned_df)
start_rows = len(cleaned_df.columns)
print(f"Starting with {start_rows} rows, {start_col} columns")
print(f"Removing rows...")

# Keep Ground, Home Delivery
services_kept = ["Ground", "Home Delivery"]
cleaned_df_1 = cleaned_df[cleaned_df['std_service_type'].isin(services_kept)]

service_type_rows = len(cleaned_df_1) 
print(f"std_service_type = Ground/Home Delivery : Removed {start_rows - service_type_rows} rows")

# Keep rows with delivery time
cleaned_df_1 = cleaned_df_1.dropna(subset=['delivery_time'])
delivery_time_rows = len(cleaned_df_1) 
print(f"delivery_time = NA : Removed {service_type_rows - delivery_time_rows} rows")

# Remove rows with delivery time = 23:59:59
cleaned_df_1 = cleaned_df_1[~cleaned_df_1['delivery_time'].isin(['23:59:59'])]
delivery_time_rows_2 = len(cleaned_df_1) 
print(f"delivery_time = 23:59:59 : Removed {delivery_time_rows - delivery_time_rows_2} rows")

# Remove rows with delivery time = 00:12:00
cleaned_df_1 = cleaned_df_1[~cleaned_df_1['delivery_time'].isin(['00:12:00'])]
delivery_time_rows_3 = len(cleaned_df_1) 
print(f"delivery_time = 00:12:00 : Removed {delivery_time_rows_2 - delivery_time_rows_3} rows")

# Remove rows with no sender or recipient zip
cleaned_df_1 = cleaned_df_1[~cleaned_df_1['recipient_zip'].isin([''])]
cleaned_df_1 = cleaned_df_1[~cleaned_df_1['sender_zip'].isin([''])]
zip_rows = len(cleaned_df_1)
print(f"recipient_zip/sender_zip = '' : Removed {delivery_time_rows_3 - zip_rows} rows")

end_rows = len(cleaned_df_1)
end_col = len(cleaned_df_1.columns)
print(f"Ending with {end_rows} rows, {end_col} columns")
del cleaned_df
```

# 2. Clean data - add new features

```python
# 2. Add features
cleaned_df_2 = cleaned_df_1.copy(deep=False)
start_rows = len(cleaned_df_2)
start_col = len(cleaned_df_2.columns)
print(f"Starting with {start_rows} rows, {start_col} columns")

# Create 3-digit sender/recipient zipcode prefix
cleaned_df_2['sender_zip_prefix'] = cleaned_df_2['sender_zip'].str[:3]
cleaned_df_2['recipient_zip_prefix'] = cleaned_df_2['recipient_zip'].str[:3]
zip_prefix_rows = len(cleaned_df_2)
zip_prefix_col = len(cleaned_df_2.columns)
print(f"Added zipcode prefix: Added {zip_prefix_rows-start_rows} rows, {zip_prefix_col-start_col} columns")

# Create std_weight (weight/package count)
cleaned_df_2['std_weight'] = cleaned_df_2['weight'] / cleaned_df_2['package_count']
std_weight_rows = len(cleaned_df_2)
std_weight_col = len(cleaned_df_2.columns)
print(f"Added std_weight i.e. weight/parcel_count: Added {std_weight_rows-zip_prefix_rows} rows, {std_weight_col-zip_prefix_col} columns")

# Create date time features
## Create week of year feature
cleaned_df_2['week_number'] = cleaned_df_2['shipment_date'].dt.week
## Create day of week feature
### The day of the week with Monday=0, Sunday=6
cleaned_df_2['day_of_week'] = cleaned_df_2['shipment_date'].dt.dayofweek
datetime_rows = len(cleaned_df_2)
datetime_col = len(cleaned_df_2.columns)
print(f"Added week_number and day_of_week: Added {datetime_rows-std_weight_rows} rows, {datetime_col-std_weight_col} columns")

end_rows = len(cleaned_df_2)
end_col = len(cleaned_df_2.columns)
print(f"Ending with {end_rows} rows, {end_col} columns")
```

# 3. Create target variable Y i.e. time window

```python
# Create target variable - time windows
cleaned_df_3 = cleaned_df_2.copy(deep=False)
start_rows = len(cleaned_df_3)
start_col = len(cleaned_df_3.columns)
print(f"Starting with {start_rows} rows, {start_col} columns")

## Calculate days in transit (exclude shipment date, holidays, weekends)
print(f"Calculating days in transit...")
### Get US federal holidays as list
from pandas.tseries.holiday import USFederalHolidayCalendar
start_date = cleaned_df_3['shipment_date'].min()
end_date = cleaned_df_3['shipment_date'].max()
calendar = USFederalHolidayCalendar()
holidays = calendar.holidays(start_date, end_date).date.tolist()

### Calculate days in transit with np.busday_count
shipment_dates = [d.date() for d in cleaned_df_3['shipment_date']]
delivery_dates = [d.date() for d in cleaned_df_3['delivery_date']]
cleaned_df_3['days_in_transit'] = np.busday_count(shipment_dates, delivery_dates, holidays = holidays) - 1 # -1 because we will add transit time
print(f"Calculated days in transit")

## Convert days in transit/delivery time to days taken (with decimals)
### e.g. if parcel reaches at 12.00pm on 2nd business day, days taken is 1.5
print(f"Converting days in transit/delivery time to days taken (float)...")
from datetime import datetime, timedelta
delivery_percentage_of_day = [(timedelta.total_seconds(d) / timedelta(days=1).total_seconds() ) for d in cleaned_df_3['delivery_time']]
cleaned_df_3['days_taken_float']= cleaned_df_3['days_in_transit'] + delivery_percentage_of_day
print(f"Converted days in transit/delivery time to days taken (float)")
```

```python
## Clean anomalous days_in_transit
### Shipments that reach in 0 business days are considered -1 days in transit
cleaned_df_3.days_in_transit.value_counts()[:15].to_frame().style.bar()
```

```python
### Keep rows from -1 to 9 days in transit
start_rows = len(cleaned_df_3)
start_col = len(cleaned_df_3.columns)
print(f"Starting with {start_rows} rows, {start_col} columns")

max_days_to_keep = 10
cleaned_df_3 = cleaned_df_3[cleaned_df_3['days_in_transit'].isin(np.arange(-1,max_days_to_keep))]
days_in_transit_rows = len(cleaned_df_3) 
print(f"Keep -1 to 9 days in transit: Removed {start_rows - days_in_transit_rows} rows")

end_rows = len(cleaned_df_3)
end_col = len(cleaned_df_3.columns)
print(f"Ending with {end_rows} rows, {end_col} columns")
```

```python
## Create time window logic
### Window percentage of day
#### 10.30am 
ten_thirty_am_percentage_of_day = timedelta(hours=10, minutes=30).total_seconds() / timedelta(days=1).total_seconds()
#### 4.30pm
four_thirty_pm_percentage_of_day = timedelta(hours=16, minutes=30).total_seconds() / timedelta(days=1).total_seconds()
print(f"10.30am: {ten_thirty_am_percentage_of_day} day, 4.30pm: {four_thirty_pm_percentage_of_day} day")
```

```python
### Create time window thresholds
#### Each day is split into 3 time windows -
#### Window 1: 12.00.01am to 10.30.00am, Window 2: 10.30.01am to 4.30.00pm, Window 3: 4.30.01pm to 12.00.00am 
time_window_thresholds = []
for i in np.arange(max_days_to_keep):
    time_window_thresholds.append(i + ten_thirty_am_percentage_of_day)
    time_window_thresholds.append(i + four_thirty_pm_percentage_of_day)
    time_window_thresholds.append(i+1)
  
### Create time window assignment function 
#### Total of 31 time windows: 
#### 0 business days -> Window 0, 
#### 3 time windows * each of 1-10 business days -> Windows 1 - 31
def assign_time_window(time): 
    # If 0 business days, assign to window 0
    if time <= 0: return 0
    lowerBound = 0
    # If >0 business days, assign to respective time window
    for upperBound in time_window_thresholds: 
        if lowerBound < time <= upperBound:
            # Use index of thresholds to assign time windows
            return time_window_thresholds.index(upperBound)+1 # +1 since Window 0 is occupied
            break
```

```python
### Assign time window 
start_rows = len(cleaned_df_3)
start_col = len(cleaned_df_3.columns)
print(f"Starting with {start_rows} rows, {start_col} columns")

cleaned_df_3['Y'] = cleaned_df_3['days_taken_float'].apply(assign_time_window)

end_rows = len(cleaned_df_3)
end_col = len(cleaned_df_3.columns)
print(f"Ending with {end_rows} rows, {end_col} columns")

### Visualize time window assignment
cleaned_df_3.Y.value_counts().to_frame().style.bar()
```

# 5. Categorize and visualize columns

```python
cleaned_df_5 = cleaned_df_4.copy(deep=False)

cat_cols = ['industry','sub_industry','shipper','std_service_type','zone',
                        'sender_state','sender_zip_prefix','recipient_state','recipient_zip_prefix',
                        'week_number','day_of_week','Y']

float_cols = ['std_weight','freight_charges']
```

```python
cleaned_df_5[cat_cols] = cleaned_df_5[cat_cols].astype('category')
cleaned_df_5[float_cols] = cleaned_df_5[float_cols].astype('float64')
```

```python
def visualize_float_columns(dataframe):
    '''
    This function visualizes Box-and-whisker plots for continuous variables
    '''
    
    # FLoat columns
    for i in float_cols:
        sns.boxplot(dataframe[i])

        # Print the three DISTINCT highest values 
        highest_vals = list(dataframe.sort_values(i, ascending=False).drop_duplicates([i], keep='last')[i].head(20))
        
        smallest_val = min(dataframe[i])
        
        plt.text(smallest_val, -0.3, highest_vals[0])
        plt.text(smallest_val, -0.2, highest_vals[10])
        plt.text(smallest_val, -0.1, highest_vals[19])

        plt.show()
```

```python
def visualize_cat_columns(dataframe):
    '''
    Lists the distinct values for categorical columns
    '''
    # Categorical columns 
    for i in cat_cols:
        print(f'Field name: {i}') # print field name
        print(f'Number of distinct values: {len(dataframe[i].unique())}') # print number of distinct values
        # for each distinct value print the number of occurances
        total = 0
        for idx, val in enumerate(dataframe[i].unique()):
            print(f'{idx+1}. {val}: {np.sum(dataframe[i] == val):,d}')
            total += np.sum(dataframe[i] == val)
        print(f'Total values: {total:,d}')
        print("---") 
```

```python
visualize_float_columns(cleaned_df_5)
visualize_cat_columns(cleaned_df_5)
```
